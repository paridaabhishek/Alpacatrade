```python
df = pd.read_csv('/content/drive//MyDrive/AlpecaTrading/DataFiles/StreamFiles/AAPL.csv',skiprows=3,index_col=False)
```

```python
df.columns =['Message Type', 
                'Symbol', 
                'Open', 
                'Close',
                'High',
                'Low',
                'Volume',
                'TimeStamp',
                'n',
                'vw']
```

```python
df
```

```python
df_unique=df.drop_duplicates(subset=['TimeStamp']).copy()
```

```python
df_unique.head(15)
```

```python
df_unique.loc[:,'TimeStamp'] = pd.to_datetime(df_unique['TimeStamp']) 
```

```python
df_unique.set_index('TimeStamp',inplace=True)
```

```python
df_unique['5MinReminder'] = df_unique.index.minute % 5
df_unique['15MinReminder'] = df_unique.index.minute % 15
```

```python
df_unique.head(20)
```

```python
df_unique.loc[df_unique[df_unique['5MinReminder']==0].index.min()]
```

```python
dtime=df_unique[df_unique['5MinReminder']==0].index.min()
dtime
```

```python
from datetime import datetime
from datetime import timedelta
```

```python
final_time = dtime + timedelta(minutes=1)
```

```python
 df_unique[df_unique['5MinReminder']==0].index.min() + timedelta(minutes=1)
```

```python
final_time
```

```python
df_unique_start=df_unique.loc[df_unique[df_unique['5MinReminder']==0].index.min() + timedelta(minutes=1):].copy()
```

```python
t=df_unique_start.groupby(pd.Grouper(freq='5min')).agg({'Open':'first',
                                                       'Close':'last',
                                                       'High':'max',
                                                       'Low':'min'}).shift(1).iloc[1:]
```
